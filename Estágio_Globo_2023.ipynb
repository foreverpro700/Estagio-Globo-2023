```python
#Importando Biblioteca Pandas como pd
import pandas as pd
```

```python
# Leitura dos dados
df = pd.read_csv("Pasta1.csv", sep=";", decimal=",", encoding="UTF-8")

# Convertendo a coluna de data para o formato correto
df['data'] = pd.to_datetime(df['data'], format='%m/%d/%Y').dt.strftime('%d/%m/%Y')

# Verificando o Data Frame
display(df)
```

```python
# Questão 1: Determine o volume de horas consumidas por categoria
volume_por_categoria = df.groupby('id_conteudo')['horas_consumidas'].sum().reset_index()

categoria_preferida = volume_por_categoria.loc[volume_por_categoria['horas_consumidas'].idxmax()]

categoria_preferida_ranking = volume_por_categoria.sort_values(by='horas_consumidas', ascending=False)

# Exibindo Resultados
print("Questão 1: Volume de horas consumidas por categoria:")

print(categoria_preferida_ranking)
```

```python
# Questão 2: Qual conteúdo teve maior volume de usuários únicos consumindo
usuarios_por_conteudo = df.groupby('id_conteudo')['id_user'].nunique().reset_index()

conteudo_mais_consumido = usuarios_por_conteudo.loc[usuarios_por_conteudo['id_user'].idxmax()]

conteudo_mais_consumido_ranking = usuarios_por_conteudo.sort_values(by='id_user', ascending=False)

print("\nQuestão 2: Volume de usuários únicos consumindo:")

print(conteudo_mais_consumido_ranking)
```

```python
# Questão 3: Determine minutos por play de cada conteúdo

# Covertendo Horas em Minutos
df['minutos_por_play'] = df['horas_consumidas'] * 60

engajamento_por_conteudo = df.groupby('id_conteudo')['minutos_por_play'].mean().reset_index()

conteudo_mais_engajado = engajamento_por_conteudo.loc[engajamento_por_conteudo['minutos_por_play'].idxmax()]

conteudo_mais_engajado_ranking = engajamento_por_conteudo.sort_values(by='minutos_por_play', ascending=False)

print("\nQuestão 3: Por minutos por play de cada conteúdo:")

print(conteudo_mais_engajado_ranking)
```
